```python


# Grouping the teams based on the "Home Team" column and summing up the home goals and home shots
grouped_data = df.groupby("Home Team").agg(
    {"Goals Home": 'sum', "home_shots": 'sum'})

# Team with the most home goals
most_goals_team = grouped_data["Goals Home"].idxmax()
most_goals_count = grouped_data["Goals Home"].max()

# Team with the most home shots
most_shots_team = grouped_data["home_shots"].idxmax()
most_shots_count = grouped_data["home_shots"].max()

# Printing the results
print("Team with the most home goals:", most_goals_team)
print("Number of goals:", most_goals_count)

print("\nTeam with the most home shots:", most_shots_team)
print("Number of shots:", most_shots_count)
```

```python
# Ranking of teams according to goal matches
df_sorted = df.sort_values(by="Goals Home", ascending=False)

# Information of the team that scored the most goals in the home team
home_team_to_score_the_most_goals = df_sorted.iloc[4]["Home Team"]
top_goals_scored_at_home = df_sorted.iloc[5]["Goals Home"]


# Graph creation
plt.figure(figsize=(10, 6))
plt.bar(df_sorted["Home Team"], df_sorted["Goals Home"])
plt.xlabel("Soccer Teams")
plt.ylabel("Number of home goals")
plt.title("Premier 2023 - Number of Goals Scored by Home Teams")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.annotate(f"Home Team with Most Goals: {home_team_to_score_the_most_goals}\nNumber of Goals: {top_goals_scored_at_home}",
             xy=(0, top_goals_scored_at_home), xytext=(0.5, top_goals_scored_at_home + 1),
             arrowprops=dict(facecolor='black', shrink=0.05))


plt.tight_layout()
plt.show()
```

```python

# List of teams
teams = df["Home Team"]

# Home and away goals
home_goals = df["Goals Home"]
away_goals = df["Away Goals"]

# Colors for scatter plot
colors = None

# Graph creation
plt.figure(figsize=(10, 6))
plt.scatter(teams, home_goals, s=200, c=colors,
            marker='o', label="Home Number of Goals")
plt.scatter(teams, away_goals, s=200, c=colors,
            marker='x', label='Deplasman Gol Sayısı')
plt.xlabel("Teams")
plt.ylabel("Number of Goals")
plt.title("Premier 2023 - Teams' Home and Away Goals")
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# List of teams
teams = df["Home Team"]

# Total number of goals
total_goals = df["Goals Home"] + df["Away Goals"]

# Graph creation
plt.figure(figsize=(10, 6))
plt.hist(total_goals, bins=10, color="blue", edgecolor="black")
plt.xlabel("Football teams")
plt.ylabel("Total number of goals")
plt.title("Premier 2023 - Total Goals of Teams")
plt.xticks(range(min(total_goals), max(total_goals)+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

# Let's calculate the total number of passes to the teams in their own court
team_passes = df.groupby('Home Team')['home_pass'].sum().reset_index()

# Let's use bar chart for visualization
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Home Team', y='home_pass', data=team_passes)
plt.xlabel("Soccer teams")
plt.ylabel("Total Number of Passes")
plt.title("The Team with the Most Passing at Home")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.show()
```

```python
# Let's calculate the total number of passes to the teams in their own court
team_passes = df.groupby('Home Team')['home_corners'].sum().reset_index()

# Let's use bar chart for visualization
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Home Team', y='home_corners',
                 color="red", data=team_passes)
plt.xlabel("Soccer teams")
plt.ylabel("Home team most corners")
plt.title("Teams' Home Total Corners Graph")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.show()
```
